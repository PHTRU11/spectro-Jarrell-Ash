```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from haggis.math import full_width_half_max
import matplotlib.transforms as tfrms
```

```python

#resolution théorique selon theta

w = 25e-3
d = 1/590
d2 = 1/50
lamb = 633e-6
m = 2
m2 = 16
L = 102
r = 2000
theta = np.linspace(0, np.pi/2, 100)
dlamb = np.sqrt(((w*d*np.cos(theta))/(m*r))**2+ ((lamb*d)/(m*L))**2)*1e6
dlamb2 = np.sqrt(((w*d2*np.cos(theta))/(m2*r))**2+ ((lamb*d2)/(m2*L))**2)*1e6
plt.plot(theta, dlamb,'k', label= '590/mm, 1$\mu m$, ordre 2')
plt.plot(theta, dlamb2,'r', label= '50/mm, 10$\mu m$, ordre 16')
plt.xlabel('largeur fente (mm)')
plt.ylabel('résolution (nm)')
plt.legend()
plt.savefig('res theo.jpg', dpi = 900)
plt.show()

```

```python
#resolution théorique selon lambda

w = 25e-3
d = 1/590
d2 = 1/50
lamb = np.arange(0, 1.5e-3,1e-6)

m = 2
m2 = 16
L = 102
r = 2000
theta = 0.3827505065110969
dlamb = np.sqrt(((w*d*np.cos(theta))/(m*r))**2+ ((lamb*d)/(m*L))**2)*1e6
dlamb2 = np.sqrt(((w*d2*np.cos(theta))/(m2*r))**2+ ((lamb*d2)/(m2*L))**2)*1e6
lamb = lamb*1e6
plt.plot(lamb, dlamb,'k', label= '590/mm, 1$\mu m$, ordre 2')
plt.plot(lamb, dlamb2,'r', label= '50/mm, 10$\mu m$, ordre 16')
plt.xlabel('largeur fente (mm)')
plt.ylabel('résolution (nm)')
plt.legend()
plt.savefig('res theo lambda.jpg', dpi = 900)
plt.show()
```
